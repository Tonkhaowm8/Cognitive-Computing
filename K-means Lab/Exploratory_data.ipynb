# Exploratory Data Analysis

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

Step 1: Load the given Banknote authentication dataset

```python
dts = pd.read_csv('bankNoteAuth.csv')
#print(dts.to_string())
```

```python
# convert to numpy
arr = dts.to_numpy()
print(arr[0][0])
```

Step 2: Calculate statistical measures, e.g. mean and standard deviation. 

```python
arr_mean = arr.mean()
arr_std = np.std(arr)
axis_mean = arr.mean(axis = 0)
print("axis mean: " + str(axis_mean))
print("Mean: " + str(arr_mean))
print("Standard Deviation: " + str(arr_std))
```

Step 3: Visualise your data as you considered fit

```python
plt.scatter(dts["V1"], dts["V2"], edgecolors='black')
plt.scatter(axis_mean[0], axis_mean[1], edgecolors='black')
plt.show()
```
